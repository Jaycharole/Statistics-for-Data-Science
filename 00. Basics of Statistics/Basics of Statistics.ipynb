# Introduction to Statistics

The first the first step of any statistical analysis is to know whether the data you are dealing is a population or sample.

# Population:
Collection of all items of interest is known as population. A population is the entire pool from which statistical sample is drawn. 

It is denoted by N. The number obtained in population is known as parameters.

# Sample:
A subset of the population is known as sample. It is called the representative of population. Sample refers to set of observations drawn from population. 

It is denoted by n. The number obtained in sample is called statistics.

# Types of Statistics

There are two types of statistics:



1. Descriptive statistics
2. Inferential statistics



## Descriptive statistics:

Descriptive Statistics is that branch of statistics which is concerned with describing the population under study. It organize, analyze and present data in a meaningful way. It is used to describe the situation.
It explains the data which is already known, to summarize people.

## Inferential Statistics:

Inferential Statistics is a type of statistics, that focuses on drawing conclusions about the population, on the basis of sample analysis and observation. It compares, test and predicts data. It explain the chances of occurence of an event. It attempts to reach the conclusion to learn about the population, that extends beyond the data available.
